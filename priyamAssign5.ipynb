```python
import numpy as np
initials='A B'
ascii_sum=sum(ord(char) for char in initials if char.isalpha())
sales=np.array([ascii_sum+i*50 for i in range(5)])
print(sales)
```

```python
tax_rate=((ascii_sum%5)+5)/100
tax_amount=sales*tax_rate
print(tax_rate)
print(tax_amount)
```

```python
discount_sales=np.where(sales<(ascii_sum+100),sales*0.95,sales*0.90)
print(discount_sales)
```

```python
weekly_sales=np.vstack([discount_sales]*3)
weeks=np.arange(3).reshape(3,1)
weekly_sales=weekly_sales*(1+weeks*0.02)
print(weekly_sales)
```

```python
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
print(x)
```

```python
y1=x**2
print(y1)
```

```python
y2=np.sin(x)
print(x)
```

```python
y3=np.exp(x)
print(y3)
```

```python
y4=np.log(np.abs(x)+1)
print(y4)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1,label=r'$y=x^2$',color='r')
```

```python
plt.plot(x,y2,label=r'$y=\sin(x)$',color='g')
```

```python
plt.plot(x,y3,label=r'$y=e^x$',color='b')
```

```python
plt.plot(x,y4,label=r'$y\log(|x|+1)$',color='m')
```

```python
plt.title("Plot of Different Functions")
plt.xlabel('x values')
plt.ylabel('y values')
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.grid(True,linestyle='--',alpha=0.6)
plt.legend()
plt.show()
```
