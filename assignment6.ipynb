```python
#Q2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "Subjects": ["Math", "Science", "English", "History", "Art"],
    "Scores": [85, 90, 78, 88, 95]
}
df = pd.DataFrame(data)
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
ax = sns.barplot(x="Subjects", y="Scores", data=df, palette="viridis")
```

```python
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')
```

```python
plt.title("Subject Scores", fontsize=14)
```

```python
plt.xlabel("Subjects", fontsize=12)
```

```python

plt.ylabel("Scores", fontsize=12)
```

```python
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
plt.show()
```

```python
#Q3
import numpy as np
import matplotlib.pyplot as plt
roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
data = np.random.randn(50)
cumulative_sum = np.cumsum(data)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
```

```python
axes[0, 0].plot(cumulative_sum, color='blue', linestyle='-', marker='o')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True)
```

```python
axes[0, 1].scatter(range(50), data, color='red', alpha=0.6)
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Value")
axes[0, 1].grid(True)
```

```python
axes[1, 0].axis("off")
axes[1, 1].axis("off")
```

```python
plt.tight_layout()
plt.show()
```

```python
#q4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    "month_number": list(range(1, 13)),
    "facecream": [2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900],
    "facewash": [1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760],
    "toothpaste": [5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400],
    "bathingsoap": [9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400],
    "shampoo": [1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800],
    "moisturizer": [1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760],
    "total_units": [21100, 18330, 22470, 22270, 20960, 20140, 29550, 36140, 23400, 26670, 41280, 30020],
    "total_profit": [211000, 183300, 224700, 222700, 209600, 201400, 295500, 361400, 234000, 266700, 412800, 300200]
}
df = pd.DataFrame(data)
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker='o', color='b', label='Total Profit')
```

```python
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker='o', color='b', label='Total Profit')
plt.title("Total Profit Over Months")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.legend()
plt.grid(True)
plt.show()
```

```python
top_products = ["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]
plt.figure(figsize=(12, 8))
for product in top_products:
    sns.lineplot(x=df["month_number"], y=df[product], marker='o', label=product)
```

```python
for product in top_products:
    sns.lineplot(x=df["month_number"], y=df[product], marker='o', label=product)
plt.title("Monthly Sales of Different Products")
plt.xlabel("Month Number")
plt.ylabel("Sales Units")
plt.legend(title="Products")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_sum = df[top_products].sum()
sns.barplot(x=df_sum.index, y=df_sum.values, palette='viridis')
```

```python
sns.barplot(x=df_sum.index, y=df_sum.values, palette='viridis')
plt.title("Total Annual Sales of Each Product")
plt.xlabel("Product")
plt.ylabel("Total Sales Units")
plt.grid(axis='y')
plt.show()
```
