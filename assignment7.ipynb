```python
#q1 part 2nd
roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
sales_data = np.random.randint(1000, 5001, size=(12, 4))
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
df = pd.DataFrame(sales_data, columns=columns, index=months)
print("First 5 Rows:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())
```

```python
total_sales_per_category = df.sum()
total_sales_per_month = df.sum(axis=1)
print("\nTotal Sales Per Category:")
print(total_sales_per_category)
print("\nTotal Sales Per Month:")
print(total_sales_per_month)
```

```python
growth_rates = df.pct_change().mean() * 100
print("\nAverage Sales Growth Between Consecutive Months (%):")
print(growth_rates)
```

```python
df["Total Sales"] = total_sales_per_month
df["Growth Rate"] = df["Total Sales"].pct_change() * 100
if roll_number % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.9
    print("\n10% discount applied to Electronics category.")

print("\nUpdated Sales Data:")
print(df)

```

```python
#q1 part 3
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
for column in columns:
    plt.plot(df.index, df[column], marker='o', label=column)
plt.title("Monthly Sales Trends for Each Category")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend()
plt.grid(True)
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[columns])
plt.title("Sales Distribution for Each Category")
plt.xlabel("Product Category")
plt.ylabel("Sales Units")
plt.grid(True)
plt.show()
```

```python
#q4
def swapElements(lst,index1,index2):
    if index1<len(lst) and index2<len(lst):
        temp=lst[index1]
        lst[index1]=lst[index2]
        lst[index2]=temp
    else:
        print("Invalid")

l1=[5,4,2,7,10]
print(l1)
swapElements(l1,1,4)
print(l1)
```

```python
#q5
def swapElementsSet(s,elem1,elem2):
    lst=list(s)
    if elem1 in lst and elem2 in lst:
        index1=lst.index(elem1)
        index2=lst.index(elem2)
        lst[index1]=lst[index2]
        lst[index2]=lst[index1]
        return set(lst)
    else:
        print("element not found in list")
        return s

my_set = {10, 20, 30, 40, 50}
print(my_set)
print(swapElementsSet(my_set,10,40))
        
    
```
